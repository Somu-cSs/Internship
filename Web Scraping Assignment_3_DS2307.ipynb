# Q2. In the above question, now scrape the following details of each product listed in first 3 pages of your search 
results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then 
scrape all the products available under that product name. Details to be scraped are: "Brand 
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and 
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
# Create an empty list to store the URLs
urls = []

# Define the number of pages you want to scrape (adjust as needed)
num_pages_to_scrape = 3

for _ in range(num_pages_to_scrape):
    # Find all the links on the current page
    page_links = driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    
    # Extract the URLs from the links and add them to the 'urls' list
    for link in page_links:
        url = link.get_attribute("href")
        if url:
            urls.append(url)
    
    # Check if there's a "Next" button and click it to go to the next page
    next_button = driver.find_element(By.XPATH, '//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    if next_button:
        next_button.click()
        time.sleep(3)  # Add a delay to allow the next page to load
    else:
        break  # Exit the loop if there is no "Next" button

# Print or process the collected URLs as needed
for url in urls:
    print(url)
```

# Q3. Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
import time
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By

# Connecting to the webdriver
driver = webdriver.Chrome()

# Getting the webpage of the mentioned URL
url = "http://images.google.com/"
driver.get(url)

# Creating empty lists
urls = []

search_item = ["Fruits", "Cars", "Machine Learning", "Guitar", "Cakes"]

for item in search_item:
    driver.get(url)
    time.sleep(5)

    # Finding web element for the search bar
    search_bar = driver.find_element(By.NAME, "q")

    # Sending keys to get the keyword for the search bar
    search_bar.send_keys(str(item))

    # Clicking on the search button
    search_button = driver.find_element(By.XPATH, "//button[@type='submit']")
    search_button.click()

    # Scrolling down the webpage to get more images
    for _ in range(10):
        driver.execute_script("window.scrollBy(0,1000)")
        time.sleep(2)

    imgs = driver.find_elements(By.XPATH, "//img[@class='rg_i Q4LuWd']")
    
    img_url = []
    for image in imgs:
        source = image.get_attribute('src')
        if source is not None and source[:4] == 'http':
            img_url.append(source)
    
    for i in img_url[:10]:
        urls.append(i)

# the directory to save the images

save_dir ="C:/Users/somas/Downloads/downloaded_images/"

for i in range(len(urls)):
    if i >= 50:  # Limit the number of images to download
        break
    
    response = requests.get(urls[i])
    
    with open(save_dir + "image_" + str(i) + ".jpg", "wb") as file:
        file.write(response.content)

# Close the WebDriver
driver.quit()
```

```python
# 1000 times we scroll down by 10000 in order to generate more comments
for _ in range(1000):
    driver.execute_script("window.scrollBy(0,10000)")
```

```python
# creating empty lists
comments = []
comment_time = []
Likes = []
No_of_Likes = []

# scrape comments
cm = driver.find_elements(By.ID, "content-text")
for i in cm:
    if i.text is None:
        comments.append("--")
    else:
        comments.append(i.text)

# scrape time when comment was posted
tm = driver.find_elements(By.XPATH, "//a[contains(text(),'ago')]")


# scrape the comment likes
like = driver.find_elements(By.XPATH,"//span[@class='style-scope ytd-comment-action-buttons-renderer']")
for i in like:
    Likes.append(i.text)
    
for i in range(1,len(Likes),2):
    No_of_Likes.append(Likes[i])
# Iterate through comments, time, and likes simultaneously
for i in range(len(cm)):
    if i < len(tm):
        comment_time.append(tm[i].text)
    else:
        comment_time.append("--")  # If there is no corresponding time, add a placeholder

    if i < len(like):
        Likes.append(like[i].text)
    else:
        Likes.append("--")  # If there is no corresponding like, add a placeholder

# Now, we have comments, corresponding times, and likes in synchronized lists
```

```python
# Create a DataFrame
Youtube = pd.DataFrame({
    'Comment': comments[:500],
    'Comment Time': comment_time[:500],
    'Comment Upvotes': No_of_Likes[:500]
})
print(Youtube)

# Save the DataFrame to a CSV file
Youtube.to_csv('youtube_comments.csv', index=False)
```

```python
# creating empty lists to store data
hostel_name = []
distance = []
rating = []
reviews = []
over_all = []
url_list= []

ele = driver.find_elements(By.XPATH, '//div[@class="property-card"]')

for index, _ in enumerate(ele):
    # Scraping hostel name
    try:
        name = _.find_element(By.XPATH, './/div[@class="property-name"]/span')
        current_hostel_name = name.text
    except NoSuchElementException:
        current_hostel_name = '-'
    hostel_name.append(current_hostel_name)

    # Scraping distance
    try:
        dist = _.find_element(By.XPATH, './/span[@class="distance-description"]')
        current_distance = dist.text
    except NoSuchElementException:
        current_distance = '-'
    distance.append(current_distance)

    # Scraping rating
    try:
        rate = _.find_element(By.XPATH,'.//span[@class="left-margin"]')
        current_rating = rate.text
    except NoSuchElementException:
        current_rating = '-'
    rating.append(current_rating)

    # Scraping overall reviews
    try:
        o_r = _.find_element(By.XPATH, './/span[@class="keyword"]')
        current_overall = o_r.text
    except NoSuchElementException:
        current_overall = '-'
    over_all.append(current_overall)

    # Scraping total reviews
    try:
        rev = _.find_element(By.XPATH, f'(.//span[@class="number"])')
        current_reviews = rev.text
    except NoSuchElementException:
        current_reviews = '-'
    reviews.append(current_reviews)
    
    # Scraping URL
    try:
        hostel_url = _.find_element(By.XPATH, './/a[@class="nuxt-link"]')
        current_url = hostel_url.get_attribute("href")
    except NoSuchElementException:
        current_url = '-'
    url_list.append(current_url)

```

```python
# next page
next_button = driver.find_element(By.XPATH, '//button[@class="pill-content page-nav nav-right icon-only"]')
    
# Check if the button is disabled (indicating no more pages)
if next_button.get_property('disabled'):
    pass

# Click on the next button
next_button.click()

# Again 

ele = driver.find_elements(By.XPATH, '//div[@class="property-card"]')

for index, _ in enumerate(ele):
    # Scraping hostel name
    try:
        name = _.find_element(By.XPATH, './/div[@class="property-name"]/span')
        current_hostel_name = name.text
    except NoSuchElementException:
        current_hostel_name = '-'
    hostel_name.append(current_hostel_name)

    # Scraping distance
    try:
        dist = _.find_element(By.XPATH, './/span[@class="distance-description"]')
        current_distance = dist.text
    except NoSuchElementException:
        current_distance = '-'
    distance.append(current_distance)

    # Scraping rating
    try:
        rate = _.find_element(By.XPATH,'.//span[@class="left-margin"]')
        current_rating = rate.text
    except NoSuchElementException:
        current_rating = '-'
    rating.append(current_rating)

    # Scraping overall reviews
    try:
        o_r = _.find_element(By.XPATH, './/span[@class="keyword"]')
        current_overall = o_r.text
    except NoSuchElementException:
        current_overall = '-'
    over_all.append(current_overall)

    # Scraping total reviews
    try:
        rev = _.find_element(By.XPATH, f'(.//span[@class="number"])')
        current_reviews = rev.text
    except NoSuchElementException:
        current_reviews = '-'
    reviews.append(current_reviews)
    
    # Scraping URL
    try:
        hostel_url = _.find_element(By.XPATH, './/a[@class="nuxt-link"]')
        current_url = hostel_url.get_attribute("href")
    except NoSuchElementException:
        current_url = '-'
    url_list.append(current_url)
```

```python
print(len(hostel_name),len(distance),len(rating),len(reviews),len(over_all),len(url_list))
```

```python
description = []
facilities = []

# Scraping for Property description And Facilities by going through each URL
for j in url_list:
    if j == '-':
        description.append('-')
        facilities.append('-')
    else:
        driver.get(j)
        time.sleep(3)
    
        # Property description
        try:
            description_xpath = '//div[@class="description-container"]//div[@class="content"]'
            description_element = driver.find_element(By.XPATH, description_xpath)
            description_text = description_element.text
        except NoSuchElementException:
            description_text = '-'

        description.append(description_text)
        
        # facilites
        try:
            # XPath expression to locate the <ul> element
            ul_xpath = '//ul[@class="facilities"]'

            # Find the <ul> element using the XPath
            ul_element = driver.find_element(By.XPATH, ul_xpath)

            # Extract the texts from all <li> elements within the <ul> element
            li_elements = ul_element.find_elements(By.XPATH, './li')
            facility_texts = [li.text.strip() for li in li_elements]

            facilities.append(facility_texts)
            
        except NoSuchElementException:
            facilities.append('-')

```

```python
print(len(description),len(facilities))
```

```python
df = pd.DataFrame({
    'hostel_name': hostel_name,
    'distance': distance,'rating':rating,'reviews':reviews ,'over all':over_all,'URLS':url_list,'Facilites':facilities,'Description':description})

# Remove the first row (index 0) from the DataFrame
df = df.drop(0)

# Reset the index after removing the row
df = df.reset_index(drop=True)

# Save the DataFrame to a CSV file
df.to_csv('hostel_details.csv', index=False)
df
```
